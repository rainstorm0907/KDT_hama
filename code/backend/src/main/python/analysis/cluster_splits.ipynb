```python
from pathlib import Path

import pandas as pd
```

```python
csv_path = Path("results/bracket_clusters/bracket_cluster_detail_20260522_1123.csv")
df = pd.read_csv(csv_path, encoding="utf-8-sig")

print(f"rows: {len(df):,}")
df.head()
```

```python
import re
from collections import Counter

SPECIAL_OR_SPACE_PATTERN = re.compile(r"[^\w가-힣]+")


def split_to_min_tokens(value):
    normalized = SPECIAL_OR_SPACE_PATTERN.sub(" ", str(value))
    return [token for token in normalized.split(" ") if token]


token_counter = Counter()
for value in df["split_tokens"].dropna():
    token_counter.update(split_to_min_tokens(value))

split_token_counts_df = pd.DataFrame(
    token_counter.most_common(),
    columns=["token", "count"],
)

split_token_counts_df.head(30)
```

```python
split_token_counts_df.head(50)
```

```python
labeling_output_dir = Path("results/bracket_clusters")
labeling_output_dir.mkdir(parents=True, exist_ok=True)

split_token_labeling_df = split_token_counts_df.assign(
    blacklist_label="",
    blacklist_reason="",
    note="",
)

labeling_csv_path = labeling_output_dir / "split_token_counts_for_blacklist_labeling.csv"
split_token_labeling_df.to_csv(labeling_csv_path, index=False, encoding="utf-8-sig")

print(f"saved: {labeling_csv_path}")
print(f"rows: {len(split_token_labeling_df):,}")
split_token_labeling_df.head(30)
```
