```python
from pathlib import Path

import pandas as pd
```

```python
csv_path = Path("results/bracket_clusters/bracket_cluster_detail_20260522_1123.csv")
df = pd.read_csv(csv_path, encoding="utf-8-sig")

print(f"rows: {len(df):,}")
df.head()
```

```python
import re
from collections import Counter

SPECIAL_OR_SPACE_PATTERN = re.compile(r"[^\w가-힣]+")


def split_to_min_tokens(value):
    normalized = SPECIAL_OR_SPACE_PATTERN.sub(" ", str(value))
    return [token for token in normalized.split(" ") if token]


token_counter = Counter()
for value in df["split_tokens"].dropna():
    token_counter.update(split_to_min_tokens(value))

split_token_counts_df = pd.DataFrame(
    token_counter.most_common(),
    columns=["token", "count"],
)

split_token_counts_df.head(30)
```
